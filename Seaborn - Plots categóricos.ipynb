# Seaborn - plots categóricos

```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Barplot

```python
sns.barplot(x='sex', y='total_bill', data=tips) 
```

Mostra média e traço preto indica desvio padrão

Passando função:

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator = np.std) 
```

### Countplot

Mostra quantidade

```python
sns.countplot(x='sex', data=tips) 
```

### boxplot

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

Hue quebra boxplot em 2:

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex') 
```

Utilizando o boxplot horizontal, se pega somente dados numéricos (não categóricos):

```python
sns.boxplot(data=tips, palette='rainbow', orient='h') 
```

### Violinplot

O violinplot permite ver a distribuição dos dados

```python
 sns.violinplot(x='day', y='tip', data=tips) 
```

Utilizando hue:

```python
 sns.violinplot(x='day', y='tip', data=tips, hue='sex') 
```

Tirando o espelhamento para despoluir:

```python
 sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True) 
```

### Stripplot

Plota pontos onde dados aparecem:

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

Utilizando o hue:

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
```

Utilizando o dodge para separar os pontos de ambos os sexos:

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

### Swarmplot

Coloca um do lado do outro a medida em que vão aparecendo:

```python
sns.swarmplot(x='day', y='total_bill', data=tips) 
```

O Problema do swarmplot é que com grande quantidade de dados, os pontos começam a se sobrepor. Não fica bom de visualizar

Utilizando o hue:

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
```

Separando com dodge:

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

Plotando swarmplot e violinplot juntos para uma boa visualização e compreensão da distribuição:

```python
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
sns.violinplot(x='day', y='total_bill', data=tips)
```

### factorplot

Assume qualquer forma a partir do kind:

```python
sns.factorplot(x='sex', y='total_bill', data=tips, kind='bar')
```
