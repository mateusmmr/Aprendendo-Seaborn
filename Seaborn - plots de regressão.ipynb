# Seaborn - plots de regress√£o

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### lmplot (Linear Model)

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

Dividindo por sexo:

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

Acrescentando marcadores diferentes:

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=["o","v"])
```

Como o lmplot √© construido em cima do scatterplot, pode-se mudar parametros do scatter a partir do lm:

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=["o","v"], scatter_kws={'s':100}) # Aumentando tamanho dos marcadores
```

Segregando dados por sexo dos clientes:

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col = 'sex')
```

Segregando por sexo do cliente e por turno:

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col = 'sex', row='time')
```

```python
Customizando o tamanho dos gr√°ficos:
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col = 'day', aspect=0.6, height=8)
```
