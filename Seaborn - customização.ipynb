# Seaborn - customização

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Importando dataset 'tips' do Seaborn:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

Customização do fundo:

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
```

Retirando as bordas com despine:

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine() # retira bordas
```

Customizando tamanho:

```python
sns.set_style('ticks')
plt.figure(figsize=(12,4))
sns.countplot(x='sex', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, height=2, aspect=4)
```

Customizando tamanho do gráfico e do nome dos eixos:

```python
sns.set_context('poster')
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.set_context('poster', font_scale=2)
sns.lmplot(x='total_bill', y='tip', data=tips)
```

Customizando esquema de cores: (Categorias disponíveis em: https://matplotlib.org/3.2.0/tutorials/colors/colormaps.html)

```python
sns.set_context('poster', font_scale=1)
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.set_context('poster', font_scale=1)
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm')
```

```python
sns.set_context('poster', font_scale=1)
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='cividis')
```

```python
sns.set_context('poster', font_scale=1)
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='GnBu')
```
