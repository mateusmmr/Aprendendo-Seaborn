# Seaborn - Plots de distribuições

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

### Distplot:

```python
sns.distplot(tips['total_bill'], kde=False) #KDE = Kernel Density Estimation
```

### Joinplot

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex') #kind - tipo ref, hex, scatter
```

### Pairplot

```python
sns.pairplot(tips, hue='sex') # Pega dados numéricos e plota vários joinplots e distplots hue = cores diferentes pra dados diferentes palette = palheta de cores
```

### Rugplot

```python
sns.rugplot(tips['total_bill']) # Pequeno traço pra cada dados - base pro kde
```

### Kdeplot

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
